```python
from requests_html import HTMLSession
import pandas as pd
import re
```

```python
session=HTMLSession()
```

```python
def get_html(url):
    res=session.get(url)
#     print(res.encoding)
    return res

# 保存图片
def get_pic_from_url(url,filename):
    pic_content = session.get(url,stream=True).content
    with open(filename,'wb') as f:
        f.write(pic_content)

#     得到网页中信息
def get_info(url):
    res=get_html(url)
    list_rank=[]
    showtime='未知'
#     css选择器
    list_res=res.html.find('body > div > div > div > div > ul > li')
    for item in list_res:
        rank=item.find('div > i', first=True).text
        title=item.find('div.txt > p.pTit> span > a ', first=True).text
        title_link=str(item.find('div.txt > p.pTit> span > a ', first=True).absolute_links).replace('{','').replace('}','').replace("'",'')
        actor=[i.text for i in item.find('div > p.pActor>a')]
#         print('  '.join(actor))
        actor_info={i.text:str(i.absolute_links).replace('{','').replace('}','').replace("'",'') for i in item.find('div > p.pActor>a')}
        actor_num=len(actor)
        if item.find('div.txt > p.pTit > span.sIntro', first=True):
            showtime=item.find('div.txt > p.pTit > span.sIntro', first=True).text.replace('上映时间：','')
        summary=item.find('div.txt > p.pTxt.pIntroShow', first=True).text.replace('简介：','')
        otherinfo=str(item.find('div.pic > img', first=True))
#         pattern=re.compile(r''' src='(.*?)'.(.*?)''',re.S)
#         pic_path='https:'+pattern.search(otherinfo).group().replace("'",'').replace(' src=','')
        pattern=re.compile(r'''//imgwx\d+.2345.com/dypcimg/img/(.*?/+.*?/+\w+).[(jpg)|[(png)]*''',re.S)
        pic_path='https:'+pattern.search(otherinfo).group()
        print(pic_path)
# 调用保存图片方法
        path='d:/book/pic/'+title.replace(':',' ')+pic_path[-4:]
        get_pic_from_url(pic_path,path)
        list_rank.append([rank,title,title_link,'  '.join(actor),summary,otherinfo,showtime,pic_path])
    return list_rank


if __name__=='__main__':
    url='https://dianying.2345.com/top/'
    data=get_info(url)
#     保存信息进excel
    df = pd.DataFrame((row  for row in get_info(url)),columns=['rank','title','title_link','actor','summary','otherinfo','showtime','pic_path'])
#     print(df)
#     df.to_excel('d:/book/moive.xlsx',index=False)






```
